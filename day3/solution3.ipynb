## Riddle 1

```python
with open("input3") as input:
  report_lines = [line.strip() for line in input.readlines()]

# Initialize histogram
histogram = []
for i in range(len(report_lines[0])):
  histogram.append({'0': 0, '1': 0})

# Fill histogram
for report_line in report_lines:
  for i in range(len(report_line)):
    histogram[i][report_line[i]] += 1

most_common = list(map(lambda e: '0' if e['0'] > e['1'] else '1', histogram))
least_common = list(map(lambda e: '0' if e['0'] <= e['1'] else '1', histogram))

gamma_rate = int("".join(most_common), 2)
epsilon_rate = int("".join(least_common), 2)

print(f'The power consumption is {gamma_rate*epsilon_rate}')
```

## Riddle 2

```python
with open("input3") as input:
  report_lines = [line.strip() for line in input.readlines()]

def create_histogram(lines):
  # Initialize histogram
  histogram = []
  for i in range(len(lines[0])):
    histogram.append({'0': 0, '1': 0})

  # Fill histogram
  for report_line in lines:
    for i in range(len(report_line)):
      histogram[i][report_line[i]] += 1
  return histogram

def filter_lines(lines, position, value):
  print(f'Before filtering:')
  print("\n".join(lines))
  filtered = list(filter(lambda line: line[position] == value, lines))
  print(f'After filtering:')
  print("\n".join(filtered))
  return filtered

# TODO regenerate histogram for each step
# oxygen generator rating
oxygen_lines = list(report_lines)
for idx in range(len(oxygen_lines[0])):
  histogram = create_histogram(oxygen_lines)
  hist_entry = histogram[idx]
  if hist_entry['1'] >= hist_entry['0']:
    oxygen_lines = filter_lines(oxygen_lines, idx, '1')
  else:
    oxygen_lines = filter_lines(oxygen_lines, idx, '0')
  if len(oxygen_lines) == 1:
    break

oxygen_generator_rating = int("".join(oxygen_lines[0]), 2)

# CO2 scrubber rating
co2_lines = list(report_lines)
for idx in range(len(co2_lines[0])):
  histogram = create_histogram(co2_lines)
  hist_entry = histogram[idx]
  if hist_entry['1'] < hist_entry['0']:
    co2_lines = filter_lines(co2_lines, idx, '1')
  else:
    co2_lines = filter_lines(co2_lines, idx, '0')
  if len(co2_lines) == 1:
    break

co2_scrubber_rating = int("".join(co2_lines[0]), 2)
print(f'Oxygen rating: {oxygen_generator_rating}')
print(f'CO2 rating: {co2_scrubber_rating}')
print(f'The life support rating is {co2_scrubber_rating*oxygen_generator_rating}')
```
